# 3b

```python
from networkx.algorithms.community import greedy_modularity_communities
```

```python
split1 = np.array_split(G.nodes, 5)
```

```python
print(split1)
```

```python
for n in split1:
    for i in n:
        for j in n:
            G.add_edge(i,j)
            
```

```python
nx.draw(G)
```

```python
uu=greedy_modularity_communities(G, weight=None)
```

```python
color_map = []
for node in G:
    if node in uu[0] :
        color_map.append('mistyrose')
    elif node in uu[1] :
        color_map.append('chocolate')
    elif node in uu[2] :
        color_map.append('cyan')
    elif node in uu[3] :
        color_map.append('purple')
    elif node in uu[4] :
        color_map.append('magenta')     
nx.draw(G, node_color=color_map)
plt.show()
```
